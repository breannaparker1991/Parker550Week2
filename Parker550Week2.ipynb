```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

this dataset has 71 years of NBA stats. columns are by players and there are 20+ stats
I would like to know who played in the NBA for the longest about of time

```python
df = pd.read_excel("NBA_Stats.xlsx")
```

```python
df.columns = df.columns.str.replace('Minutes Played', 'min_play')
df.columns = df.columns.str.replace('Season Start Year', 'season_year')
df.columns = df.columns.str.replace('Games Played', 'games_played')
```

```python
#Checking to see if there are any nan values for the columns I'm using along with how the columns are named
for c in df.columns:
    nan = df[c].isnull().values.any()
    #printing out 
    if nan == True:
        print(" {} has Nan value(s)".format(c)) 
    else: 
        print("{} has NO Nan values!".format(c))
```

```python
#showing the total minutes played in their entire career(so far)
df4 = df.groupby('Player')['min_play'].sum().nlargest(10)
df4.head(10)
```

```python
df4 = pd.DataFrame({'player': ['Kareem Abdul-Jabbar', 'LeBron James', 'Karl Malone', 'Kobe Bryant', 'Dirk Nowitzki', 'Tim Duncan', 'Jason Kidd',
                               'Kevin Garnett', 'Wilt Chamberlain',   'John Stockton' 
                                ], 'total_min': [66297, 65746, 62759, 57284, 57262, 56737, 56203, 55695, 55418, 54164]}, dtype=object)
df4
```

```python
#graphing the total number played by player in total
fig, ax = plt.subplots(figsize=(15, 5))
ax.scatter(df4['player'], df4['total_min'], s=50, facecolor='C0', edgecolor='k')
ax.set_xlabel('Player Names')
ax.set_ylabel('Minutes Played Total')
```

```python
df1 = df.sort_values(by=['min_play'])
```

```python
df1=df1.tail(20)
df1
```

```python
#this figure shows the top 20 times for most minutes played in a respective season
fig, ax = plt.subplots(figsize=(15, 5))
ax.scatter(df1['Player'], df1['min_play'], s=50, facecolor='C0', edgecolor='k')
ax.set_xlabel('Player Names')
ax.set_ylabel('Minutes Played Per Season')
```

The above graph shows us that player Wilt Chamberlain, not only has the most played minutes per season played overall. But has also played the most minutes per season for 10 out of the 20 top minutes played in a season. 

```python
#counting how many seasons each player is shown by rows
df2 = df.groupby(["Player"])["season_year"].count().nlargest(10)
df2
```

```python
#converting the count into a new df with easily usable columns
df2 = pd.DataFrame({'player': ['Eddie Johnson', 'John Stockton', 'Kareem Abdul-Jabbar', 'Karl Malone', 'Mike Dunleavy', 'Robert Parish', 'Tim Duncan', 'Dirk Nowitzki', 'Jason Kidd', 'LeBron James'], 'count': [42, 38, 38, 38, 38, 37, 37, 36, 36, 36]}, dtype=object)
df2
```

```python
#plotting the count of how many seasons played by player
fig, ax = plt.subplots(figsize=(15, 10), layout='constrained')
ax.bar(df2['player'], df2['count'])
plt.xlabel('Player names')
plt.ylabel('number of seasons played')
```

The above graph shows the total number of seasons that each of the longest player played. We see that the person who played the most seasons was Eddie Johnson

```python
#sorting by most games played, top 10
df3 = df.sort_values(by=['games_played'])
df3 = df3.tail(10)
df3
```

```python
#graph showing the top ten most games played per season. 
plt.figure(figsize=(15,6))
plt.plot(df3['Player'],df3['games_played'], linewidth=5.0)
plt.ylabel('games played per season')
plt.xlabel('player names')
plt.show()
```

The above graph shows the most games played per season by player. We can see that the most games played in a season is tied by Theo and Chris

I originally wanted to know which player played for the longest in the NBA. However, after looking at the data, I realized that this questions could be answered in several different ways. In my first graph we show the total minutes played in their entire career, which is Kareem Abdul-Jabbar. However, this data may not be a good reflection of NBA players who are still in the middle or beginning of their career. So, then we look at the player who has the most minutes played in any given season, which was Wilt Chamberlain. I decided to also look at the total number of seasons played, which is also not completely accurate due to the number of players who are still playing. But this name was Eddie Johnson. Lastly, I decided to look at the most number of games played in a given season, thinking that this would correlate with the most minutes played, but then a new name appeared as Theo Ratliff. So, there could have been players that played in every game, but only played for a few minutes, which was different than the players that played for a large part of a given game, but was not in every game. You would also think thta the player who has played the longest would have the most minutes played overall, but if a single player only played a little per game and kept doing that for several seasons, that would allow them to have played the most seasons without having played the most minutes overall. This was a very interesting experiment on how important it is to look at multiple data points when finding your answer.
